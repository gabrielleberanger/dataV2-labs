```python
# your code here
employees[['Years','Salary']].groupby('Years').mean()
```

```python
# your code here
employees[['Department','Salary']].groupby('Department').agg(lambda x: max(x)-min(x))
```

#### Select all the customers who have bought more than 50 items.
**Hint**: Remember that you are not asked to find orders with more than 50 items but customers who bought more than 50 items. A customer with two orders of 30 items each should appear in the selection.

```python
# your code here
df = orders[['CustomerID','Quantity']].groupby('CustomerID').sum().reset_index()
df.loc[df.Quantity>50,'CustomerID'].values
```

```python
# your code here
df = orders[['InvoiceNo','Quantity']].groupby('InvoiceNo').sum().reset_index()
df.loc[df.Quantity>50,'InvoiceNo'].values
```
